# Gr√°ficos dos agentes

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Single Agent - Ambiente 1c

```python
df_tradicional_1c = pd.read_csv("../results/traditional_search_1_agent_1c.csv")
df_expanding_1c = pd.read_csv("../results/expanding_search_1_agent_1c.csv")
df_AStar_1c = pd.read_csv("../results/AStar_search_1_agent_1c.csv")

plt.figure(figsize=(8, 5))
plt.plot(df_tradicional_1c["step"], df_tradicional_1c["accumulated_pos"], label="Traditional Search", color='blue')
plt.plot(df_expanding_1c["step"], df_expanding_1c["accumulated_pos"], label="Expanding Search", color='green')
plt.plot(df_AStar_1c["step"], df_AStar_1c["accumulated_pos"], label="A* Search", color='red')

plt.xlabel("Steps", fontsize=14, fontweight='bold')
plt.ylabel("Accumulated Position", fontsize=14, fontweight='bold')
plt.title("Single Agent - Ambiente 1c ", fontsize=16, fontweight='bold')

plt.xticks(ticks=range(0, df_tradicional_1c["step"].max() + 1, 10))
plt.yticks(fontsize=12)
plt.grid(True, linestyle="--", alpha=0.7)

plt.legend(title="Search Algorithms", fontsize=12, title_fontsize=13, loc="best")

plt.tight_layout()
plt.show()
```
