# Gr√°ficos dos agentes

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Tradicional Maritime Search Single Agent

```python
df_teadicional_1c = pd.read_csv("../results/traditional_search_1_agent_1c.csv")
df_teadicional_2c = pd.read_csv("../results/traditional_search_1_agent_2c.csv")

plt.figure(figsize=(8, 5))
plt.plot(df_teadicional_1c["step"], df_teadicional_1c["accumulated_pos"])
plt.plot(df_teadicional_2c["step"], df_teadicional_2c["accumulated_pos"])

plt.xlabel("Steps", fontsize=14, fontweight='bold')
plt.ylabel("Accumulated Position", fontsize=14, fontweight='bold')
plt.title("Tradicional Maritime Search ", fontsize=16, fontweight='bold')

plt.xticks(ticks=range(0, df_teadicional_1c["step"].max() + 1, 10))
plt.yticks(fontsize=12)

plt.grid(True, linestyle="--", alpha=0.7)
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()
```
